# Survivor Data Exploration

Explore the **survivorpy** package — a Python wrapper for data from the R package [survivoR](https://github.com/doehm/survivoR). The data covers contestants, seasons, episodes, votes, challenges, and more from the reality show *Survivor*.

**Note:** On first import, survivorpy fetches and caches data locally (may take a moment). Subsequent loads are fast and work offline.

## 1. Setup & Import

Install survivorpy if needed, then import and load the package.

```python
# Uncomment if survivorpy is not installed:
# !pip install survivorpy

import survivorpy as sv
import pandas as pd
from IPython.display import display

print(f"survivorpy loaded. Last synced: {sv.LAST_SYNCED}")
print(f"\nAvailable tables: {len(sv.TABLE_NAMES)}")
```

## 2. List All Available Tables

```python
for i, name in enumerate(sv.TABLE_NAMES, 1):
    print(f"{i:2}. {name}")
```

## 3. Explore Each Table

For each table: shape, column names, dtypes, and a preview of the first few rows.

```python
def explore_table(name: str) -> None:
    """Load a table and print its structure and sample."""
    try:
        df = sv.load(name)
        print(f"\n{'='*60}")
        print(f"TABLE: {name}")
        print(f"{'='*60}")
        print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
        print(f"\nColumns ({len(df.columns)}):")
        for col in df.columns:
            dtype = str(df[col].dtype)
            print(f"  • {col}: {dtype}")
        print(f"\nFirst 5 rows:")
        display(df.head())
    except Exception as e:
        print(f"\n[Error loading '{name}']: {e}")
```

```python
# Explore all tables (run each to see output)
for table_name in sv.TABLE_NAMES:
    explore_table(table_name)
```

## 4. Quick Summary: Table Sizes & Key Columns

```python
summary = []
for name in sv.TABLE_NAMES:
    try:
        df = sv.load(name)
        summary.append({
            "Table": name,
            "Rows": df.shape[0],
            "Columns": df.shape[1],
            "Column Names": ", ".join(df.columns[:8].tolist()) + ("..." if len(df.columns) > 8 else "")
        })
    except Exception as e:
        summary.append({"Table": name, "Rows": "—", "Columns": "—", "Column Names": str(e)})

display(pd.DataFrame(summary))
```

## 5. Sample: Castaways & Seasons

A closer look at two core tables — castaways (contestants) and seasons.

```python
castaways = sv.load('castaways')
print("Castaways table:")
castaways.info()
display(castaways.head(10))
```

```python
if 'season_summary' in sv.TABLE_NAMES:
    seasons = sv.load('season_summary')
    print("Season summary:")
    display(seasons.head(15))
elif 'seasons' in sv.TABLE_NAMES:
    seasons = sv.load('seasons')
    print("Seasons:")
    display(seasons.head(15))
else:
    print("Season info may be in castaways or another table.")
    print("Unique season IDs in castaways:", castaways['season'].nunique() if 'season' in castaways.columns else "N/A")
```
